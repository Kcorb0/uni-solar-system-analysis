```python
import requests
import json
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Configs
pd.set_option('display.max_columns', None)
```

```python
def create_dataframe(url):
    re = requests.get(url)
    bodies = json.loads(re.text)["bodies"]
    df = pd.DataFrame(bodies)
    return df
```

```python
url = "https://api.le-systeme-solaire.net/rest/bodies"
bodies_df = create_dataframe(url)
```
